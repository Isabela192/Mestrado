```python
import pyart
import matplotlib.pyplot as plt
```

```python
colors = ['BlueBrown10','BlueBrown11','BrBu10','BrBu12','Bu10','Bu7','BuDOr12','BuDOr18','BuDRd12','BuDRd18',
          'BuGr14','BuGy8','BuOr10','BuOr12','BuOr8','BuOrR14','Carbone11','Carbone17',
          'Carbone42','Cat12','EWilson17','GrMg16','Gray5','Gray9','NWSRef','NWSVel',
          'NWS_SPW','PD17','RRate11','RdYlBu11b','RefDiff','SCook18','StepSeq25','SymGray12',
          'Theodore16','Wild25','LangRainbow12']
colors_2 = ['pyart_' + color for color in colors]
radar = pyart.aux_io.read_gamic('./Sao_Roque/2017-02-02/RADL01046020170202191024.mvol')
display = pyart.graph.RadarDisplay(radar)
```

```python
rows = 7
cols = 5
f = plt.figure(figsize = [20,25])
for i, vals in enumerate(colors_2):    
    plt.subplot(rows,cols, i+1)
    display.plot('reflectivity', sweep=1, vmin=--10, vmax=70, cmap=vals, title=vals)
    plt.tight_layout()
    plt.savefig('sample')
```
